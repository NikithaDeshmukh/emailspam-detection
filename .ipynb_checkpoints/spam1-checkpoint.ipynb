# Data Preprocessing
-Lower case
-Tokenization
-Removing special characters
-Removing stop words and punctuation

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import string

nltk.download('stopwords')  # You have a typo in the download function call

ps = PorterStemmer()  # Fix the typo in the class instantiation

def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)  # Fix the indentation of this line
    text = y[:]  # Fix the indentation of this line
    y.clear()
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:  # Fix the syntax of this line
            y.append(i)
    text = y[:]
    y.clear()
    for i in text:
        y.append(ps.stem(i))
    return " ".join(y)

transformed_text = transform_text("I am gonna be home soon and I don't want to talk this stuff anymore tonight, k? I have cried today")
print(transformed_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd

# Load or create your DataFrame, specifying the encoding
df = pd.read_csv(r'C:\Users\nikitha deshmukh\OneDrive\Desktop\EMAIL-CLASSIFICATION\spam.csv', encoding='latin1')

# Assuming 'v1' contains your target variable
X = df['v2']  # Assuming 'v2' contains your text data column
y = df['v1']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# Assuming X_train and X_test are text data
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Assuming y_train and y_test are string labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Create and train Gaussian Naive Bayes classifier
gnb = GaussianNB()
gnb.fit(X_train_tfidf.toarray(), y_train_encoded)
y_pred_gnb = gnb.predict(X_test_tfidf.toarray())

# Create and train Multinomial Naive Bayes classifier
mnb = MultinomialNB()
mnb.fit(X_train_tfidf, y_train_encoded)
y_pred_mnb = mnb.predict(X_test_tfidf)

# Create and train Bernoulli Naive Bayes classifier
bnb = BernoulliNB()
bnb.fit(X_train_tfidf, y_train_encoded)
y_pred_bnb = bnb.predict(X_test_tfidf)

# Evaluate Gaussian Naive Bayes
print("Gaussian Naive Bayes:")
print("Accuracy:", accuracy_score(y_test_encoded, y_pred_gnb))
print("Confusion Matrix:\n", confusion_matrix(y_test_encoded, y_pred_gnb))
print("Precision Score:", precision_score(y_test_encoded, y_pred_gnb))

# Evaluate Multinomial Naive Bayes
print("\nMultinomial Naive Bayes:")
print("Accuracy:", accuracy_score(y_test_encoded, y_pred_mnb))
print("Confusion Matrix:\n", confusion_matrix(y_test_encoded, y_pred_mnb))
print("Precision Score:", precision_score(y_test_encoded, y_pred_mnb))

# Evaluate Bernoulli Naive Bayes
print("\nBernoulli Naive Bayes:")
print("Accuracy:", accuracy_score(y_test_encoded, y_pred_bnb))
print("Confusion Matrix:\n", confusion_matrix(y_test_encoded, y_pred_bnb))
print("Precision Score:", precision_score(y_test_encoded, y_pred_bnb))
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score
from sklearn.preprocessing import LabelEncoder

accuracy_scores = []
precision_scores = []
algorithm_names = []

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit the LabelEncoder with the training target data
label_encoder.fit(y_train)

for name, clf in clfs.items():
    # Train the classifier
    clf.fit(X_train, y_train_encoded)
    
    # Make predictions
    y_pred_encoded = clf.predict(X_test)
    
    # Decode the predictions
    y_pred = label_encoder.inverse_transform(y_pred_encoded)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)
    
    # Calculate precision
    precision = precision_score(y_test, y_pred, pos_label='spam')
    precision_scores.append(precision)
    
    # Store algorithm name
    algorithm_names.append(name)

# Create a DataFrame to store the results
performance_df = pd.DataFrame({
    'Algorithm': algorithm_names,
    'Accuracy': accuracy_scores,
    'Precision': precision_scores
})

```

```python
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

X_train_text = ["Sample text 1", "Sample text 2", "Sample text 3"]
y_train = [0, 1, 0]  # Assuming you have labels for your training data

tfidf = TfidfVectorizer(lowercase=True, stop_words='english')
X_train_tfidf = tfidf.fit_transform(X_train_text)

mnb = MultinomialNB()
mnb.fit(X_train_tfidf, y_train)

with open('Vectorizer.pkl', 'wb') as vectorizer_file:
    pickle.dump(tfidf, vectorizer_file)

with open('model.pkl', 'wb') as model_file:
    pickle.dump(mnb, model_file)
```
